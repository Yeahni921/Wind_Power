```python
import os
import pandas as pd
import glob

base_dir1 = '/home/user/Wind_Power/notebooks/TEST_bidirectional_mse'
base_dir2 = '/home/user/Wind_Power/notebooks/TEST_bidirectional_smoothL1'

csv_files1 = glob.glob(os.path.join(base_dir1, '**', '*.csv'), recursive=True)
csv_files2 = glob.glob(os.path.join(base_dir2, '**', '*.csv'), recursive=True)
```

```python
# total results (mse loss)
mse_result_df = []

for csv_file in csv_files1:
    df = pd.read_csv(csv_file)
    
    min_val_row = df.loc[df['Val Loss'].idxmin()] 
    model_name = os.path.basename(os.path.dirname(csv_file))
    
    mse_result_df.append({
        'model_name': model_name,
        'min_val_loss': min_val_row['Val Loss'],
        'MAE': min_val_row['MAE'],
        'gj_NMAE' : min_val_row['MAE']/20700,
        'yg_NMAE' : min_val_row['MAE']/79600,
        'criterion' : 'MSE'
    })

mse_result_df = pd.DataFrame(mse_result_df)
```

```python
mse_result_df.sort_values(by='MAE')
```
